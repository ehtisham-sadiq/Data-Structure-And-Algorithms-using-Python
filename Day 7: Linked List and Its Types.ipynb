```python
import networkx as nx
import matplotlib.pyplot as plt

class Node:
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next_node = next_node

class LinkedList:
    def __init__(self):
        self.head = None
        
    def add_node(self, value):
        new_node = Node(value)
        new_node.next_node = self.head
        self.head = new_node
        
    def visualize(self):
        G = nx.DiGraph()
        current_node = self.head
        while current_node:
            G.add_node(current_node)
            if current_node.next_node:
                G.add_edge(current_node, current_node.next_node)
            current_node = current_node.next_node
        nx.draw(G, with_labels=False)
        plt.show()

linked_list = LinkedList()
linked_list.add_node(1)
linked_list.add_node(2)
linked_list.add_node(3)
linked_list.visualize()
```

```python
from graphviz import Digraph

class Node:
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next_node = next_node

class CircularLinkedList:
    def __init__(self):
        self.head = None
        
    def add_node(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            new_node.next_node = self.head
        else:
            current_node = self.head
            while current_node.next_node != self.head:
                current_node = current_node.next_node
            current_node.next_node = new_node
            new_node.next_node = self.head
    
    def delete_node(self, value):
        current_node = self.head
        previous_node = None
        while current_node.next_node != self.head:
            if current_node.value == value:
                if previous_node:
                    previous_node.next_node = current_node.next_node
                else:
                    self.head = current_node.next_node
                return
            previous_node = current_node
            current_node = current_node.next_node
        if current_node.value == value: # This is for the case where the head node is being deleted
            previous_node.next_node = current_node.next_node
            self.head = current_node.next_node
            
    def search_node(self, value):
        current_node = self.head
        while current_node.next_node != self.head:
            if current_node.value == value:
                return current_node
            current_node = current_node.next_node
        if current_node.value == value: # This is for the case where the head node is being searched
            return current_node
        return None

    def visualize(self):
        dot = Digraph()
        current_node = self.head
        while current_node.next_node != self.head:
            dot.node(str(current_node), str(current_node.value))
            dot.edge(str(current_node), str(current_node.next_node))
            current_node = current_node.next_node
        dot.node(str(current_node), str(current_node.value))
        dot.edge(str(current_node), str(self.head))
        dot.render('circular_linked_list.gv', view=True)
        
circular_linked_list = CircularLinkedList()
circular_linked_list.add_node(31)
circular_linked_list.add_node(24)
circular_linked_list.add_node(37)
circular_linked_list.add_node(13)
circular_linked_list.add_node(42)
circular_linked_list.add_node(83)
circular_linked_list.visualize()
circular_linked_list.delete_node(13)
circular_linked_list.visualize()
node = circular_linked_list.search_node(3)
if node:
    print("Node found: ", node.value)
else:
    print("Node not found")
```
